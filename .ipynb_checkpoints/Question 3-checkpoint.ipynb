```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('Data/FPL_tweets.csv', delimiter=',')

data['Interaction'] = data['Likes'] + data['Replies'] + data['Retweets']
sorted_by_interaction = data.sort_values(by='Interaction', ascending=False)
```

```python
colors = plt.cm.Set3(np.linspace(0, 1, 3))
plt.figure(figsize=(12, 6))
```

```python
for i, (likes, retweets, replies) in enumerate(zip(sorted_by_interaction['Likes'][:10], sorted_by_interaction['Retweets'][:10], sorted_by_interaction['Replies'][:10])):
    plt.bar(sorted_by_interaction['User'][i], likes, color=colors[0], label='Likes' if i == 0 else "")
    plt.bar(sorted_by_interaction['User'][i], retweets, bottom=likes, color=colors[1], label='Retweets' if i == 0 else "")
    plt.bar(sorted_by_interaction['User'][i], replies, bottom=likes+retweets, color=colors[2], label='Replies' if i == 0 else "")

plt.xlabel('User')
plt.ylabel('Total Interaction')
plt.title('Top 10 Tweets with Most Interaction')

legend_labels = ['Likes', 'Retweets', 'Replies']
legend_handles = [plt.Rectangle((0,0),1,1, color=colors[i]) for i in range(len(legend_labels))]
plt.legend(legend_handles, legend_labels, title='Interactions', loc='upper right')

plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
